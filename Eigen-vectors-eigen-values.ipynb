```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Create a dataset

```python
x = np.linspace(-3, 3, 1000)
y = np.sin(x) + np.random.randn(1000)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(which='both')
plt.title("A 2D dataset")
plt.show()
```

## Calculate covariance of y and x

```python
xy = np.column_stack((x,y))
xy_mean_zeroed = xy - np.mean(xy)
xy_mean_zeroed = xy
cov_xy = np.cov(xy.T)
eigenValues, eigenVectors = np.linalg.eig(cov_xy)

print(f"Eigen vectors: \n{eigenVectors}")
print(f"Corresponding eigen values: \n{eigenValues}")
```

## Plot x y and eigen vectors

```python
plt.figure()
plt.plot(xy_mean_zeroed[:, 0], xy_mean_zeroed[:, 1], 'o')
plt.quiver(0, 0, eigenVectors[0][0], eigenVectors[1][0], scale=np.sqrt(5*np.max(eigenValues)/eigenValues[0]), zorder=100, angles='xy', color='red')
plt.quiver(0, 0, eigenVectors[0][1], eigenVectors[1][1], scale=np.sqrt(5*np.max(eigenValues)/eigenValues[1]), zorder=100, angles='xy', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(which='both')
plt.title("A 2D dataset")
plt.show()
```

## Calculate principal components of x y

```python
xy_principal_components = np.matmul(xy, eigenVectors)
cov_xy_principal_components = np.cov(xy_principal_components.T)
eigenValues_pc, eigenVectors_pc = np.linalg.eig(cov_xy_principal_components)

plt.figure()
plt.plot(xy_principal_components[:, 0], xy_principal_components[:, 1], 'o')
plt.quiver(0, 0, eigenVectors_pc[0][0], eigenVectors_pc[1][0], scale=np.sqrt(5*np.max(eigenValues)/eigenValues[0]), zorder=100, angles='xy', color='red')
plt.quiver(0, 0, eigenVectors_pc[0][1], eigenVectors_pc[1][1], scale=np.sqrt(5*np.max(eigenValues)/eigenValues[1]), zorder=100, angles='xy', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(which='both')
plt.title("Principal components of the 2D dataset")
plt.show()
```
